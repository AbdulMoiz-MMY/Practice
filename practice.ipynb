```python
model = genai.GenerativeModel(
    'gemini-1.5-flash',
    generation_config=genai.GenerationConfig(
        max_output_tokens=450,
        temperature=1
    )
)
```

```python
response = model.generate_content("What is AI")
```

```python
print(response.text)
```
